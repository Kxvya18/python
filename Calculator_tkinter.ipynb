```python
import tkinter as tk
from tkinter import ttk
import sv_ttk

window = tk.Tk()
window.geometry("350x550")
window.title("Calculator")

# Set the theme
sv_ttk.set_theme("dark")

expression = ''

def test(n):
    global expression
    expression += str(n)
    equation.set(expression)

def out():
    global expression
    try:
        output = str(eval(expression))
        equation.set(output)
        expression = ''
    except:
        equation.set('Error')
        expression = ''

def clear():
    global expression
    expression = ''
    equation.set('0')

def back():
    global expression
    expression = expression.rstrip(expression[-1])
    equation.set(expression)

exp_frame = ttk.Frame(window, style="TFrame")
button_frame = ttk.Frame(window, style="TFrame")

exp_frame.pack()
button_frame.pack()

font_spec = ('ariel', 20, 'bold')
font_button = ('ariel', 12)

equation = tk.StringVar()
equation.set('0')

input_exp = ttk.Entry(exp_frame, textvariable=equation, font=font_spec, justify='left')
input_exp.pack(ipadx=10, ipady=20, pady=(10))

button1 = ttk.Button(button_frame, text='1', style="TButton", command=lambda: test(1))
button2 = ttk.Button(button_frame, text='2', style="TButton", command=lambda: test(2))
button3 = ttk.Button(button_frame, text='3', style="TButton", command=lambda: test(3))
button_plus = ttk.Button(button_frame, text='+', style="TButton", command=lambda: test('+'))

button4 = ttk.Button(button_frame, text='4', style="TButton", command=lambda: test(4))
button5 = ttk.Button(button_frame, text='5', style="TButton", command=lambda: test(5))
button6 = ttk.Button(button_frame, text='6', style="TButton", command=lambda: test(6))
button_minus = ttk.Button(button_frame, text='-', style="TButton", command=lambda: test('-'))

button7 = ttk.Button(button_frame, text='7', style="TButton", command=lambda: test(7))
button8 = ttk.Button(button_frame, text='8', style="TButton", command=lambda: test(8))
button9 = ttk.Button(button_frame, text='9', style="TButton", command=lambda: test(9))
multiply = ttk.Button(button_frame, text='*', style="TButton", command=lambda: test('*'))

button0 = ttk.Button(button_frame, text='0', style="TButton", command=lambda: test(0))
deci = ttk.Button(button_frame, text='.', style="TButton", command=lambda: test('.'))
clear_button = ttk.Button(button_frame, text='C', style="TButton", command=clear)
div = ttk.Button(button_frame, text='/', style="TButton", command=lambda: test('/'))

equal = ttk.Button(button_frame, text='=', style="TButton", command=out)
back_button = ttk.Button(button_frame, text='<<', style="TButton", command=back)
openn = ttk.Button(button_frame, text='(', style="TButton", command=lambda: test('('))
close = ttk.Button(button_frame, text=')', style="TButton", command=lambda: test(')'))

button1.grid(row=0, column=0,ipadx=20, ipady=20)
button2.grid(row=0, column=1,ipadx=20, ipady=20)
button3.grid(row=0, column=2,ipadx=20, ipady=20)
button_plus.grid(row=0, column=3,ipadx=20, ipady=20)

button4.grid(row=1, column=0,ipadx=20, ipady=20)
button5.grid(row=1, column=1,ipadx=20, ipady=20)
button6.grid(row=1, column=2,ipadx=20, ipady=20)
button_minus.grid(row=1, column=3,ipadx=20, ipady=20)

button7.grid(row=2, column=0,ipadx=20, ipady=20)
button8.grid(row=2, column=1,ipadx=20, ipady=20)
button9.grid(row=2, column=2,ipadx=20, ipady=20)
multiply.grid(row=2, column=3,ipadx=20, ipady=20)

button0.grid(row=3, column=0,ipadx=20, ipady=20)
deci.grid(row=3, column=1,ipadx=20, ipady=20)
clear_button.grid(row=3, column=2,ipadx=20, ipady=20)
div.grid(row=3, column=3,ipadx=20, ipady=20)

equal.grid(row=5, column=0, columnspan=4, sticky='nsew',ipadx=20, ipady=20)
back_button.grid(row=4, column=0, columnspan=2, sticky='nsew',ipadx=20, ipady=20)
openn.grid(row=4, column=2,ipadx=20, ipady=20)
close.grid(row=4, column=3,ipadx=20, ipady=20)

window.mainloop()
```
